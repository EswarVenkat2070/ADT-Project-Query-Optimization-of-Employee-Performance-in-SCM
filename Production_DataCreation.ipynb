```python
import pandas as pd
import datetime
import random
from google.colab import files
```

```python
def make_workers(num):
    
    # lists to randomly assign to workers
    # status_list = ['Full Time', 'Part Time']
    # products = ['Bagle', 'Donut', 'Coffee', 'wedges', 'Soups']
    Elist = ["1001", "1002", "1003","1004","1005","1006","1007","1008","1009","1010"]
    Plist = ["a1001","b1002","c1003","d1004","e1005","f1006","g1007","h1008","i1009","j1010"]
    start_date = datetime.date(2010, 1, 1)
    end_date = datetime.date(2020, 1, 1)
    delta = datetime.timedelta(days=1)
    k=[]

    while start_date <= end_date:
      # for i in range(0,7):
      #   print(Elist[i])
      fake_workers = [{'Produced Date' : start_date, 
                        'Product ID': Plist[x],
                        'Employee ID': Elist[x], 
                        'Produced Quantity': random.randint(60, 89),
                        'Total Quantity': random.randint(90, 100),
                        'Work Hours' : random.randint(4,8)}for x in range(num)]
      k.extend(fake_workers)
      start_date += delta
                  
    return k
```

```python
worker_df = pd.DataFrame(make_workers(10))
display(worker_df)
worker_df.to_excel('Production_Table.csv')
files.download('Production_Table.csv')
```
